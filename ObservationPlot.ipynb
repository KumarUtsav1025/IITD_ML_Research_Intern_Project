```python
import pandas as pd
import matplotlib.pyplot as plt
LR1= [93.21,96.38,99.05,99.07]
LR2= [93.06,96.28,99.16,99.18]
LR3= [88.17,89.92,92.43,92.97]
SVM1=[93.57,96.84,98.65,98.83]
SVM2 =[91.85,92.68,94.09,94.50]
x= ["PC4","PC5","PC6","PC9"]
```

```python

plotdata = pd.DataFrame({
    "LR(Mul+Saga)": LR1,
    "LR(Mul+Sag)": LR2,
    "LR(OvR+LibLin)": LR3,
    "SVM(OvR+rbf)": SVM1,
    "SVM(OvR+poly)": SVM2},
    index=x)

plotdata.plot(figsize=(8, 6))
plt.title("Accuracy Plot", fontsize=24)
plt.xlabel("Number of PC", fontsize=20)
plt.ylabel("Accuracy(%)", fontsize=20)

# Increase font size of tick labels
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xticks(rotation = 75)
plt.savefig("12.1.png", bbox_inches='tight')
plt.show()
```
